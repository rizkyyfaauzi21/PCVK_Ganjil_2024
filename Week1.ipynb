```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://d1tgyzt3mf06m9.cloudfront.net/production/media/2017/oktober/10-gundam-transfor/gundam-transformasi-1-700x700.jpg"]
# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url) #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB) #convert color to RGB
  final_frame = cv.hconcat((image, image_2)) #concatenate image
  cv2_imshow(final_frame) #show image
  print('\n')





```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2),y]=[255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
# urls = ["https://d1tgyzt3mf06m9.cloudfront.net/production/media/2017/oktober/10-gundam-transfor/gundam-transformasi-1-700x700.jpg"]
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for x in range (tinggi):
  image_4[x,int((lebar)/2)]=[255,255,255]

final_frame = cv.hconcat((image_2, image_4))
cv2_imshow(final_frame)
```

```python
# urls = ["https://d1tgyzt3mf06m9.cloudfront.net/production/media/2017/oktober/10-gundam-transfor/gundam-transformasi-1-700x700.jpg"]
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)


#Garis diagonal dari kiri ke kanan
for y in range(tinggi):
  x = int(lebar/tinggi*y)
  image_5[y, x] = [255,255,255]

#Garis diagonal dari kanan ke kiri
for y in range(tinggi):
  x = int((lebar/tinggi)*(tinggi-y-1))
  image_5[y, x] = [255,255,255]

final_frame = cv.hconcat((image_2, image_5))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_6 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#Buat garis horizontal di tengah dengan panjang tertentu
garis_tengah = tinggi/2
panjang_garis = lebar/4

titik_awal = (lebar - panjang_garis)/2
titik_akhir = titik_awal + panjang_garis

for y in range(int(titik_awal), int(titik_akhir)):
  image_6[int(garis_tengah), y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_6))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_7 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Definisikan titik awal dan titik akhir
titik_awal = 50  # Misalnya, 50 piksel dari tepi
titik_akhir = 100  # Misalnya, 200 piksel dari tepi

# Pastikan titik_awal dan titik_akhir dalam rentang dimensi gambar
titik_awal = max(0, titik_awal)
titik_akhir = min(image_7.shape[1], titik_akhir)  # Batasi hingga lebar gambar

# Gambar kotak putih
for y in range(titik_awal, titik_akhir):
    for x in range(titik_awal, titik_akhir):
        if y < image_7.shape[0] and x < image_7.shape[1]:  # Pastikan dalam batas gambar
            image_7[y, x] = [255, 255, 255]

# Menggabungkan gambar secara horizontal
final_frame = cv.hconcat([image_2, image_7])
cv2_imshow(final_frame)
```
